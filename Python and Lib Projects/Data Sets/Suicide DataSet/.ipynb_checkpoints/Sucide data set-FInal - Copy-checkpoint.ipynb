# Name of Attempted Suicides Coutries

```python
countries=data.country.unique()
print("-----------------------------------")
print("Name of Attempted Suicides Coutries")
print("-----------------------------------")

for i in countries:
    print("-",i)
```

# Total no of suicides attempted in countries

```python
print("---------------------------------")
print("Total no of suicides in countries")
print("---------------------------------")
suicides_no=data.suicides_no.sum()
print("Total Suicides : " , suicides_no)
```

# Updates years of Suicides

```python
years=data.year.unique()
for i in years:
    print(i, end=" |")
```

# Sucide attempt  1987 to 2016
   - visualize by Graph

```python
year=data.year
sucicedes=data.suicides_no
plt.bar(year,sucicedes,label="Suicides Attemps",color='m') 
plt.xlabel("Suicided Years")
plt.ylabel("Number of Attempted Suicides")
style.use('ggplot')
plt.legend()
plt.title("Suicides Attempt from 1987 to 2016 ")
```
